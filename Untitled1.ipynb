```python
import gym
import numpy as np
import random
import tensorflow as tf
import matplotlib.pyplot as plt

env = gym.make("FrozenLake-v0")

X = tf.placeholder(shape=[1, 16], dtype=tf.float32)
W = tf.Variable(tf.random_uniform([16, 4], 0, 0.01))
Q_out = tf.matmul(X, W)
predict = tf.argmax(Q_out, 1)
Label_Q_out = tf.placeholder(shape=[1, 4], dtype = tf.float32)

loss = tf.reduce_sum((Q_out - Label_Q_out)**2)
opt = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(loss)

jList = []
rList = []
y = 0.99
e = 0.1
num_of_episode = 2000
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(num_of_episode):#2000->num of episode
        print("--------------------------------------------------------------------------------------------------------")
        s = env.reset() # s   episode  0  . s  github observation  
        rAll = 0
        done = False
        j = 0
        while j < 99: # episode max 100  , Goal    done True  episode .
            j = j + 1
            print(s)
            a, allQ = sess.run([predict, Q_out], feed_dict = {X : np.identity(16)[s:s+1]})
            if(np.random.rand(1) < e):
                a[0] = env.action_space.sample()
            s_next, reward, done, _ = env.step(a[0])#observation(state), reward, done(boolean . goal   ), info
            
            Q1 = sess.run(Q_out, feed_dict = {X : np.identity(16)[s_next:s_next+1]})#actions_new_network -> [1, 4]
            
            maxQ1 = np.max(Q1)
            
            targetQ = allQ
            targetQ[0, a[0]] = reward + y*maxQ1
            
            _opt, _weight = sess.run([opt, W], feed_dict = {X : np.identity(16)[s:s+1], Label_Q_out : targetQ})
            
            
            #state ,  state  
            s = s_next
            
            rAll = rAll + reward
            if(done == True):
                e = 1.0/((i/50) + 10)
                print(s)
                break
        jList.append(j)
        rList.append(rAll)
print(" : ", sum(rList) / num_of_episode)
```
